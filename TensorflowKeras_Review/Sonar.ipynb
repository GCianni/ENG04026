```python
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout,  Flatten
from sklearn.preprocessing import LabelEncoder 
```

```python
df = pd.read_csv('dadosdesonar.csv')
df.head() 
```

```python
X_input = df.iloc[: , :-1]
y_label = df['Class'].values
```

```python
label_encoder_Y = LabelEncoder()
y_label = label_encoder_Y.fit_transform(y_label)
y_label = y_label.reshape([208, 1])
```

```python
modelo = Sequential()
modelo.add(Dense(300, input_dim=60, activation = 'relu'))
modelo.add(Dense(200, activation = 'relu'))
modelo.add(Dense(100, activation = 'relu'))
modelo.add(Dense(1, activation = 'sigmoid'))
modelo.compile (optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
modelo.fit (X_input, y_label, epochs=30)
```

```python
modelo.evaluate (X_input, y_label) 
```
